This project visualizes word frequency in Natural Language Processing (NLP) using Python.


```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
text = """
NLP helps computers understand language.
NLP helps in text analysis.
Language processing is an important part of NLP.
Text analysis helps extract information from language.
"""

print("Original Text:")
print(text)
```

```python
# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenize
tokens = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Filtered Tokens:")
print(filtered_tokens)
```

```python
word_freq = Counter(filtered_tokens)
print("Word Frequency:")
print(word_freq)
```

```python
# Select top 5 words
words, counts = zip(*word_freq.most_common(5))

plt.figure()
plt.bar(words, counts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 5 Most Frequent Words")
plt.show()
```
